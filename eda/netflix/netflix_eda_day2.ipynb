```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')
plt.title("Distribution of Movies vs TV Shows")
plt.show()
```

```python
year_counts = df['release_year'].value_counts().sort_index()

year_counts.plot()
plt.title("Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
## Key Findings

1. Movies significantly outnumber TV Shows.
2. Content production increased after 2015.
3. Director and country columns have substantial missing values.
4. Most content belongs to Movie category.

## Next Steps

- Clean missing values
- Explore country distribution
- Analyze genre trends
```
